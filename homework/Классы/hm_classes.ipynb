```python
# Задание 1
def get_name_carrency(url):
    dir_ = {}
    req = requests.get(url).json()
    for k , v in req['Valute'].items():
        dir_.setdefault(v['Name'], v['Value'])
        
    return sorted(dir_.items(), key=lambda x: x[1], reverse=True)[0]

url = 'https://www.cbr-xml-daily.ru/daily_json.js'
get_name_carrency(url)
```

```python
# Задание 2

class RateDiff(Rate):
    
    def __init__(self, diff = False, format='value'):
        super().__init__(format=format)
        self.diff = diff
        
    def make_format(self, currency): # переопределяю метод make_format
    
        crrncy = super().make_format(currency)
        if self.format == 'full':
            return crrncy # возвращаем родящий метод
        else:
            if not self.diff:
                return crrncy # возвращаем родящий метод
            else:
                dir_ = Rate('full').make_format(currency) # получаем словарь чтобы достать Value и Previous
                
                return dir_['Value'] - dir_['Previous']
```

```python
# Задание 3

class Employee:
    def __init__(self, name, seniority):
        self.name = name
        self.seniority = seniority
        
        self.grade = 1
    
    def grade_up(self):
        """Повышает уровень сотрудника"""
        self.grade += 1
    
    def publish_grade(self):
        """Публикация результатов аккредитации сотрудников"""
        print(self.name, self.grade)
    
    def check_if_it_is_time_for_upgrade(self):
        pass
    
class Developer(Employee):
    def __init__(self, name, seniority):
        super().__init__(name, seniority)
    
    def check_if_it_is_time_for_upgrade(self):
        # для каждой аккредитации увеличиваем счетчик на 1
        # пока считаем, что все разработчики проходят аккредитацию
        self.seniority += 1
        
        # условие повышения сотрудника из презентации
        if self.seniority % 5 == 0:
            self.grade_up()
        
        # публикация результатов
        return self.publish_grade()
```

```python
class Designer(Employee):
    def __init__(self, name, seniority, prizes = 2):
        super().__init__(name, seniority)
        self.prizes = prizes # начальные премии
        self.seniority += self.prizes * 2 # начальные баллы
        
        
    def check_if_it_is_time_for_upgrade(self):
        self.seniority += 2
        
        if self.seniority % 7 == 0:
            self.grade_up()
        
        # публикация результатов
        return self.publish_grade()
```

```python
## из за того что метод publish_grade вызыается как print, то печатаются дубли строк

# Тест 1: Проверка начальных параметров
designer = Designer("Анна", 0)
print(f"Имя: {designer.name}")          # Анна
print(f"Стаж: {designer.seniority}")     # должно быть 4 (2 премии * 2 балла)
print(f"Грейд: {designer.grade}")        # 1
print(f"Премии: {designer.prizes}")      # 2

# Тест 2: Первая аккредитация

designer.check_if_it_is_time_for_upgrade()
print(f"\nПосле 1-й аккредитации:")
print(f"Стаж: {designer.seniority}")     # должно быть 6
print(f"Грейд: {designer.grade}")         # 1 (6 % 7 != 0)

# Тест 3: Шестая аккредитация
for _ in range(5):
    designer.check_if_it_is_time_for_upgrade()
print(f"\nПосле 6-й аккредитации:")
print(f"Стаж: {designer.seniority}")     # должно быть 18
print(f"Грейд: {designer.grade}")         # 2 (18 % 7 == 0)
```
