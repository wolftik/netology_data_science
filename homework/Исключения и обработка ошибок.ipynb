## Задание 1
Печатные газеты использовали свой формат дат для каждого выпуска. Для каждой газеты из списка напишите формат указанной даты для перевода в объект datetime:

The Moscow Times — Wednesday, October 2, 2002

The Guardian — Friday, 11.10.13

Daily News — Thursday, 18 August 1977



```python
import datetime as dt

magazine1 = 'The Moscow Times — Wednesday, October 2, 2002'
magazine2 = 'The Guardian — Friday, 11.10.13'
magazine3 = 'Daily News — Thursday, 18 August 1977'

magazine1 = magazine1.split(' — ')
magazine2 = magazine2.split(' — ')
magazine3 = magazine3.split(' — ')

dt_magazine1 = dt.datetime.strptime(magazine1[1], '%A, %B %d, %Y')
dt_magazine2 = dt.datetime.strptime(magazine2[1], '%A, %d.%m.%y')
dt_magazine3 = dt.datetime.strptime(magazine3[1], '%A, %d %B %Y')

print(dt_magazine1)
print(dt_magazine2)
print(dt_magazine3)
```

## Задание 3
Напишите функцию date_range, которая возвращает список дат за период от start_date до end_date. Даты должны вводиться в формате YYYY-MM-DD. В случае неверного формата или при start_date > end_date должен возвращаться пустой список.


```python
def date_range(start_date, end_date):
    lst = []
    try:
        dt_start_date = dt.datetime.strptime(start_date, '%Y-%m-%d')
        dt_end_date = dt.datetime.strptime(end_date, '%Y-%m-%d')
        if dt_start_date > dt_end_date:
            return lst
        else:
            while dt_start_date<=dt_end_date:
                lst.append(dt_start_date)
                dt_start_date = dt_start_date + dt.timedelta(days=1)
            return lst
    except ValueError:
        return lst
    
    
print(date_range('2025-01-01', '2025-01-05'))
print(date_range('2025-01-10', '2025-01-05'))
print(date_range('2025-01-01', 'asdfsd'))
```
