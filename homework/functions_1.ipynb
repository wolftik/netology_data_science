# Домашнее задание к лекции «Функции»

Представьте, что вы помогаете библиотекарю автоматизировать учёт книг. Вам нужно написать программу, которая обрабатывает пользовательские команды на основе имеющихся данных.

### Исходные данные

**1. Перечень всех книг**

- `genre` — жанр книги
- `number` — уникальный номер
- `title` — название книги
- `author` — автор книги

**2. Перечень полок, на которых хранятся книги** (если книга есть в books, то она обязательно должна быть и в directories)

- ключ словаря (`'1'`, `'2'`, `'3'`) — это номер полки
- значение — список номеров книг, лежащих на соответствующей полке

### Общие требования к программе

- код должен быть грамотно декомпозирован (каждая функция отвечает за свою конкретную задачу, дублирующийся функционал переиспользуется, а его код не повторяется);
- в коде отсутствуют глобальные переменные (за исключением **books** и **directories**);
- пользовательский ввод обрабатывается в цикле **while** до тех пор, пока пользователь явно не завершит программу (вводом команды "**q**");
- код обязательно должен содержать конструкцию **match case** и функцию **setdefault()**.

````python
books = [
    {'genre': 'поэзия', 'number': '978-5-1000-1234-7', 'title': 'Евгений Онегин', 'author': 'Александр Пушкин'},
    {'genre': 'фэнтези', 'number': '88006', 'title': 'Властелин колец', 'author': 'Джон Р. Р. Толкин'},
    {'genre': 'детектив', 'number': 'D-1122', 'title': 'Безмолвный свидетель', 'author': 'Агата Кристи'}
]

directories = {
    '1': ['978-5-1000-1234-7', '88006'],
    '2': ['D-1122'],
    '3': []
}
````

### Пункт 1. Пользователь по команде "book_info" может найти название книги и её автора

````python
def book_info(book_num:str):
    for i, b in enumerate(books):
        if b.get('number')==book_num:
            print(f'Название книги: {b.get('title')} \nАвтор: {b.get('author')}')
            break
    else:
        print('Книга не найдена в базе')
            
    
# book_info('88006')
````

### Пункт 2. Пользователь по команде "shelf" может по названию книги узнать, на какой полке она хранится

Примеры работы:

```
Введите команду:
shelf

Введите название книги:
Безмолвный свидетель
```
Результат: `Книга хранится на полке: 2`

```
Введите команду:
shelf

Введите название книги:
Абвгд
```
Результат: `Книга не найдена в базе`

````python
def shelf_num(number:str):
    for k, v in directories.items():
                if number in v:
                    return k

def shelf(book_name):
    for i, b in enumerate(books):
        if b.get('title')==book_name:
            n = shelf_num(b['number'])
            if n:
                print(f'Книга хранится на полке: {n}')
            else:
                print('Полка не найдена')
            break
    else:
        print('Книга не найдена в базе')
                
        
shelf('Безмолвный свидетель')
````

````python
def all_books():
    for b in books:
        print(f'№: {b['number']}, жанр: {b['genre']}, название: {b['title']}, автор: {b['author']}, полка хранения: {shelf_num(b['number'])}')
        
all_books()
````

### Пункт 4. Пользователь по команде "add_shelf" может добавить новую полку

Примеры работы:

```
Введите команду:
add_shelf

Введите номер полки:
10
```
Результат: `Полка добавлена. Текущий перечень полок: 1, 2, 3, 10.`

```
Введите команду:
add_shelf

Введите номер полки:
1
```
Результат: `Такая полка уже существует. Текущий перечень полок: 1, 2, 3.`

````python
def add_shelf(shelf_number):
    if str(shelf_number) not in directories.keys():
        directories.setdefault(str(shelf_number),[])
        print(f'Полка добавлена. Текущий перечень полок: {", ".join(directories.keys())}.')
    else:
        print(f'Такая полка уже существует. Текущий перечень полок: {", ".join(directories.keys())}.')

add_shelf(4)
````

### Пункт 5. Пользователь по команде "del_shelf" может удалить существующую полку из данных (только если она пустая)

Примеры работы:

```
Введите команду:
del_shelf

Введите номер полки:
3
```
Результат: `Полка удалена. Текущий перечень полок: 1, 2.`

```
Введите команду:

Введите номер полки:
1
```
Результат: `На полке есть книги, удалите их перед удалением полки. Текущий перечень полок: 1, 2, 3.`

```
Введите команду:
del_shelf

Введите номер полки:
4
```
Результат: `Такой полки не существует. Текущий перечень полок: 1, 2, 3.`

````python
def del_shelf(shelf_number):
    if str(shelf_number) not in directories.keys():
        print(f'Такой полки не существует. Текущий перечень полок: {", ".join(directories.keys())}.')
    elif len(directories[str(shelf_number)])==0:
        directories.pop(str(shelf_number))
        print(f'Полка удалена. Текущий перечень полок: {", ".join(directories.keys())}.')
    else:
        print(f'На полке есть книги, удалите их перед удалением полки. Текущий перечень полок: {", ".join(directories.keys())}.')
        
        

del_shelf(4) 
````


````python
def get_not_input_empty(text):
    while True:
        value = input(text).strip()
        if value=='q':
            return None
        elif value:
            return value
        print('Ошибка: Поле не может быть пустым. Попробуйте еще раз.')

def add_book():
    number = get_not_input_empty("Введите номер книги: ")
    if number is None: return
    genre = get_not_input_empty("Введите жанр книги: ")
    if genre is None: return
    title = get_not_input_empty("Введите название книги: ")
    if title is None: return
    author = get_not_input_empty("Введите автора книги: ")
    if author is None: return
    shelf_put = get_not_input_empty("Введите полку для хранения: ")
    if shelf_put is None: return
    
    if str(shelf_put) not in directories.keys():
        print('Такой полки не существует. Добавьте полку командой add_shelf.\nТекущий список книг:')
        all_books()
        return
    
    books.append({'number':number,'genre':genre,'title':title,'author':author})
    directories[shelf_put].append(number)
    print('Книга добавлена. Текущий список книг:')
    all_books()
    
        
````

````python
def del_book(book_num):
    for i, b in enumerate(books):
        if b.get('number')==book_num:
            books.pop(i)
            for n in directories.values():
                if book_num in n:
                    n.remove(book_num)
            print('Книга удалена.\nТекущий список книг:')
            all_books()
            break
    else:
        print('Книга не найдена в базе.\nТекущий список книг:')
        all_books()
        
            

del_book('123456')
````

````python
# Итог

while True:
    command_input = input("Введите команду: ").lower()
    
    match command_input:
        case "q":
            break
        case "book_info":
            book_info(input("Введите номер книги: ").lower())
        case "shelf":
            shelf(input("Введите название книги: ").lower())
        case "all":
            all_books()
        case "add_shelf":
            add_shelf(int(input("Введите номер полки: ")))
        case "del_shelf":
            del_shelf(int(input("Введите номер полки: ")))
        case "add_book":
            add_book()
        case "del_book":
            del_book("Введите номер книги: ")
        case "move":
            pass
        case _:
            print("Неизвестная команда")
````
