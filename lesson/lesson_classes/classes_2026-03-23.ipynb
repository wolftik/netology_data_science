### Проверка валидности значений атрибутов

```python
class Prices:
    def __init__(self, price, discount):
        self.price = price
        self.discount = discount

    @property  # декоратор
    def price_w_vat(self):
        """Выполняет вычисления и возвращаемое значение пишет в атрибут calculated_attribute"""
        # посчитаем цену товара с НДС без скидки
        price_w_vat = self.price * 1.2
        return price_w_vat * (1 - self.discount)

    # как можно проверять назначаемые значения атрибутов
    # скидка всегда должна быть от 0 до 1
    @property
    def discount(self):
        return self._discount

    @discount.setter  # проверка на назначаемое значение
    def discount(self, value):
        if value < 0 or value > 1:
            raise ValueError(f'Значение скидки должно быть от 0 до 1. Получили: {value}')
        
        self._discount = value
```

```python
p = Prices(price=100, discount=10)
```

```python
p.price_w_vat
```

```python
from abc import ABC, abstractmethod
```

```python
class Report(ABC):
    @abstractmethod
    def sources(self):
        """откуда берем данные"""
        pass

    @abstractmethod
    def calculation(self):
        """Как именно эти данные обработать"""
        pass

    def write_to_db(self):
        # в методе запускается процесс вычислений и записи результатов
        pass
```

```python
class SalesReport(Report):
    def sources(self):
        print('Определяем источники данных')

    def calculation(self):
        print('Пишем код вычислений')
```

```python
API_URL = 'https://api-metrika.yandex.ru/stat/v1/data'
METRIKA_ROWS_LIMIT = 5  # max 100000

DEMO_COUNTER = 44147844
```

```python
class Metrika:
    def __init__(self, token='', counter=DEMO_COUNTER, start_date='2025-03-01', end_date='2025-03-08'):
        self.counter = counter
        self.limit = METRIKA_ROWS_LIMIT
        self.start_date = start_date
        self.end_date = end_date
        
        if token:
            self.headers = {'Authorization': 'OAuth ' + token}
        else:
            self.headers = ''  # для демо-счетчика заголовок запроса должен быть пустым
        
        self.dimensions = ''  # что мы хотим увидеть в строках отчета
        self.metrics = ''  # в столбцах
        self.report = []  # сам отчет
        
    def api_request(self, offset=1):
        """Запрос к API Метрики, возвращает словарь с отчетом"""
        params = {
            'id': self.counter,
            'date1': self.start_date,
            'date2': self.end_date,
            'metrics': self.metrics,
            'dimensions': self.dimensions,
            'limit': self.limit,
            'offset': offset,
            'accuracy': 1,
        }
        
        response = requests.get(API_URL, params=params, headers=self.headers)
        return response.json()['data']
    
    def traffic(self):
        """Отчет по посещаемости"""
        self.metrics = 'ym:s:visits'
        self.dimensions = 'ym:s:date'
        
        self.report = self.full_report()

    def browser(self):
        """Отчет по браузерам"""
        self.metrics = 'ym:s:visits'
        self.dimensions = 'ym:s:date,ym:s:browser'
        
        self.report = self.full_report()
    
    def reformat_api_report(self):
        """Упрощение отчета до списка списков"""
        reformatted_report = []

        for line in self.report:
            dimensions = [x['name'] for x in line['dimensions']]
            reformatted_report.append(dimensions + line['metrics'])

        self.report = reformatted_report
    
    def full_report(self):
        """Постраничная выгрузка из Метрики"""
        full_data = []
        offset = 1
        
        while True:
            print('Starting offset {}'.format(offset))
            data = self.api_request(offset=offset)
            full_data += data
            
            offset += self.limit
            if not data or len(data) < self.limit:
                break
        
        return full_data

    def to_csv(self, filename):
        with open(filename, 'w'):
            for line in self.report:
                f.write(...)
```
