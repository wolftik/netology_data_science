### F-строки: форматируем красиво! ✨

F-строки (форматированные строковые литералы) позволяют легко вставлять значения переменных и результаты выражений прямо внутрь строки, используя префикс `f` перед кавычками и фигурные скобки `{}`.

**Вспоминаем полезные возможности:**

*   **Простая вставка переменных:** `f"Пользователь {username} сделал {num_orders} заказов."`
*   **Форматирование чисел:**
    *   `{value:.2f}` - Округление до 2 знаков после запятой (для цен, средних).
    *   `{large_number:,}` - Добавление разделителей тысяч (для больших чисел).
    *   `{percentage:.1%}` - Отображение в виде процентов (значение умножается на 100).
*   **Выражения внутри скобок:** `f"Итого: {price * quantity}"`, `f"Имя в нижнем регистре: {name.lower()}"`

Представьте, что мы работаем над интернет-магазином. У нас есть данные о конкретном товаре, и нам нужно быстро и красиво отобразить их для пользователя или для внутреннего отчета. Все эти данные хранятся в обычных переменных.

````python
# Наши данные о товаре (представьте, что это одна запись из нашей базы данных)
product_name = "Игровой ноутбук 'Штурмовик Про'"
original_price = 129990     # Оригинальная цена
discount_percentage = 0.15  # 15% скидка
customer_rating = 4.87      # Средний рейтинг от 0 до 5
num_reviews = 2345          # Количество отзывов
stock_quantity = 15         # Количество на складе
is_limited_offer = True     # Флаг, указывающий на ограниченность предложения
````

````python
discounted_price = original_price * (1 - discount_percentage)
discount_amount = original_price * discount_percentage

print("--- 🚀 СПЕЦИАЛЬНОЕ ПРЕДЛОЖЕНИЕ! 🚀 ---")
# Заголовок: выравниваем по центру с помощью символа ^
# .upper() делает текст заглавными буквами
print(f"{product_name.upper():^40}") # :^40 - центрировать в поле шириной 40 символов
print("--------------------------------------------------")

# :.2f - ГАРАНТИРУЕТ 2 знака после запятой (даже если число целое, добавит .00)
# :, - добавляет разделители тысяч
print(f"✨ Старая цена: {original_price:,.2f} руб.")
print(f"🎉 Ваша скидка: {discount_percentage:.0%} (-{discount_amount:,.2f} руб.)")
print(f"🔥 Новая цена: {discounted_price:,.2f} руб. 🔥")

# :.1f - 1 знак после запятой для рейтинга
# :, - разделитель тысяч для количества отзывов
print(f"{'Рейтинг покупателей:':<25} {customer_rating:.1f} из 5 ({num_reviews:,} отзывов)") # <25 - выравнивание по левому краю в поле шириной 25

# тернарный оператор прямо внутри
stock_message = (f"Осталось всего {stock_quantity} шт. - спешите!"
                 if stock_quantity <= 20 else "В наличии, много!")
print(f"{'Наличие:':<25} {stock_message}")

# Еще один тернарный оператор для статуса предложения
offer_status = ("ЭТО ОГРАНИЧЕННОЕ ПРЕДЛОЖЕНИЕ!"
                if is_limited_offer else "Предложение действует постоянно.")
print(f"{'Статус акции:':<25} {offer_status}")

print("--------------------------------------------------")
print(f"{'Не упустите свой шанс!':^40}")
````

### Цикл `for`: Итерация по последовательностям

Цикл `for` – ваш основной инструмент для перебора элементов *любой* итерируемой последовательности (списков, кортежей, строк). Он проходит по каждому элементу, делая его доступным внутри тела цикла.



**Пример: сумма элементов главной диагонали "матрицы"** 📐

Представим данные в виде квадратной "матрицы" (списка списков). Нам нужно посчитать сумму элементов, стоящих на главной диагонали.


````python
data = [
    [10, 2, 3, 4],
    [4, 5, 6, 17],
    [7, 8, 20, 9],
    [3, 32, 1, 7]
]
````

````python
sum_ = 0
index = 0
for row in data:
    # print(row)
    sum_ += row[index]
    index += 1

print(sum_)
````

````python
sum_ = 0
index = -1
for row in data:
    # print(row)
    sum_ += row[index]
    index -= 1

print(sum_)
````

### Цикл `while`: повторение по условию ⏱️

В отличие от цикла `for`, который проходит по элементам последовательности, цикл `while` (англ. "пока") выполняется до тех пор, пока некоторое **условие истинно (`True`)**. Как только условие становится ложным (`False`), цикл прекращается.

**Ключевое отличие от `for`:** Цикл `while` используется тогда, когда мы *не знаем заранее*, сколько раз нужно выполнить итерацию, но знаем *условие*, при котором нужно продолжать или остановиться.




Предположим, что мы сделали вклад в банке с ежегодной капитализацией. На какую сумму вклад будет прирастать каждый год?

````python
money = 100000  # Начальная сумма денег
year = 0        # Счетчик лет
yearly_multiplier = 0.05  # Процентная ставка (5%)
money_target = money * 2  # Целевая сумма (в два раза больше)

print(f'Стартовый капитал: {money:.2f} у.е.')

while money < money_target:
    # 1. Сначала считаем, сколько составит прирост в этом году
    # (5% от текущей суммы)
    growth = money * yearly_multiplier

    # 2. Затем добавляем этот прирост к основной сумме
    money += growth

    # 3. Увеличиваем счетчик лет
    year += 1

    print(f'Год {year}: {money:.2f} у.е. → был прирост: {growth:.2f} у.е.')

print(f'\nЦель в {money_target:.2f} у.е. достигнута за {year} лет.')
print(f'Итоговый капитал: {money:.2f} у.е.')
````

Для успешного выполнения цикла условие должно быть построено таким образом, чтобы оно когда-нибудь стало ложным, то есть переменные, участвующие в условии, должны изменяться в теле цикла. Иначе мы уйдём в бесконечный цикл (infinite loop).

````python
money = 100000
year = 0
yearly_multiplier = 0.05
money_target = money * 2

print(f'Стартовый капитал: {money:.2f} у.е.')

while money < money_target:
    growth = money * yearly_multiplier
    money -= growth # здесь специально ошибся
    year += 1
    print(f'Год {year}: {money:.2f} у.е. → прирост: {growth:.2f} у.е.')

print(f'\nЦель в {money_target:.2f} у.е. достигнута за {year} лет.')
print(f'Итоговый капитал: {money:.2f} у.е.')
````

Всегда убеждайтесь, что внутри тела цикла происходит что-то, что в конечном итоге сделает `условие` ложным. Обычно это изменение переменной, которая проверяется в условии.

**Пример: Гипотеза Коллатца**

[Гипотеза Коллатца](https://ru.wikipedia.org/wiki/Гипотеза_Коллатца) гласит, что любое натуральное число сводится к единице при следующих действиях над ним:

а) если число чётное, то разделить его пополам;

б) если число нечётное, то умножить его на 3 и прибавить 1

Над вновь полученным числом вновь повторить действия `a` или `б` в зависимости от чётности числа.

Мы не знаем, сколько шагов понадобится, чтобы дойти до 1, поэтому `while` здесь идеален. Будем продолжать, *пока* число не равно 1.

````python
number = 4396874938756934769438798634976873053609375096703985670395875609827709870965709567
counter = 0
while number != 1:
    if number % 2 == 0:
        number //= 2
    else:
        number = number * 3 + 1
    # print(number)
    counter += 1

print(f'Количество итераций: {counter}')
````

### Блок `else` для циклов: если `break` не сработал

Блок `else`, написанный на том же уровне отступа, что и `for` или `while`, выполняется только в том случае, если цикл завершился **естественным образом** (т.е. были пройдены все итерации `for`, или условие `while` стало ложным), **а не был прерван** оператором `break`.


*   **Цикл `for`:** Если мы ищем элемент и находим его (`break`), блок `else` не выполняется. Если мы прошли весь список и не нашли (цикл завершился сам), `else` выполняется.
*   **Цикл `while`:** Если условие цикла стало ложным и мы вышли, `else` выполняется. Если мы вышли из цикла по `break`, `else` не выполняется.


````python
import random

number_to_guess = random.randint(1, 100)
print(number_to_guess)

tries_left = 3
print(f"\nИгра 'Угадай число' (1-100), {tries_left} попытки:")

while tries_left > 0:
    print(f"  Осталось попыток: {tries_left}")
    guess = int(input("  Ваш вариант: "))

    if guess > number_to_guess:
        print("  Загаданное число меньше.")
    elif guess < number_to_guess:
        print("  Загаданное число больше.")
    else:
        print("  Вы угадали! 🎉")
        break  # Выход из цикла по успеху

    tries_left -= 1
# Этот else сработает, если цикл завершился, т.к. tries_left стал 0
# (т.е. break НЕ сработал - пользователь не угадал)
else:
    print(f"\nПопытки закончились! Вы не угадали.")
    print(f"Было загадано число: {number_to_guess}")
````

### Вложенные циклы: когда они нужны (и нет)

Иногда для решения задачи нам нужно поместить один цикл внутрь другого. Это называется **вложенным циклом (nested loop)**.

**Как это работает:**
Для **каждой** итерации **внешнего** цикла, **внутренний** цикл выполняется **полностью** (от начала до конца).

Вложенные циклы чаще всего нужны в двух ситуациях:

1.  **Генерация комбинаций:** Когда нужно скомбинировать элементы из двух (или более) коллекций.
2.  **Обработка многомерных структур:** Когда данные организованы в виде "таблицы" (список списков) или другой вложенной структуры.

Представим, что мы составляем варианты комбо-обедов. У нас есть список основных блюд и список напитков. Мы хотим показать все возможные сочетания "блюдо + напиток".

````python
main_courses = ['Паста Карбонара', 'Стейк из лосося', 'Овощной салат "Цезарь"']
drinks = ['Лимонад', 'Кофе', 'Чай зеленый']

print("Возможные комбо-обеды 'Блюдо - Напиток':")
combo_count = 0
# Внешний цикл по основным блюдам
for course in main_courses:
    # Внутренний цикл по напиткам
    for drink in drinks:
        print(f"Комбо: {course} - {drink}")
        combo_count += 1
    print('- - - - - - - - - - - - -')

print(f"\nВсего возможных комбо-обедов: {combo_count}")
````

Представим, что у нас есть список данных о продажах, где каждая запись содержит город, категорию товара и количество (для простоты, пусть всегда 1). Нам нужно посчитать, сколько товаров каждой категории было продано в каждом городе.


**Важные моменты:**

*   **Производительность:** Помните, что количество операций во внутреннем цикле равно произведению числа итераций внешнего и внутреннего циклов (N*M). Для больших N и M это может быть медленно.
*   **Читаемость:** Глубокая вложенность (3+ уровня) сильно затрудняет чтение кода. Старайтесь избегать этого, возможно, разбивая логику на части (в будущем — на функции).
*   **`break`/`continue`:** Действуют только на **тот цикл, внутри которого они написаны**. `break` во внутреннем цикле не прервет внешний.

## Regex – завершаем основы

Мы возвращаемся к регулярным выражениям! На прошлом семинаре мы познакомились с базовым поиском (`re.search`) и заменой (`re.sub`), а также основными метасимволами и квантификаторами. Сегодня мы добавим к этому знание о функциях `re.findall()` и `re.split()`, а затем изучим  механизм **групп**, который позволяет извлекать и использовать части найденного текста.

### `re.findall()`: поиск всех совпадений

Часто нам нужно найти не первое, а **все** вхождения шаблона в тексте. Для этого идеально подходит `re.findall()`.

*   `re.findall(pattern, string)`: Находит **все непересекающиеся** совпадения `pattern` в `string`.
*   Возвращает **список строк**, где каждая строка — это найденное совпадение.
*   Если совпадений нет, возвращает **пустой список `[]`**.

````python
import re

text = "Контакты: +7-999-123-45-67, email: test@site.info. Запасной: +7-800-555-35-35."
phone_pattern = r'\+\d-\d{3}-\d{3}-\d{2}-\d{2}'

# Ищем ВСЕ телефоны
all_phones = re.findall(phone_pattern, text)

print(f"Найденные телефоны (findall): {all_phones}")
print(f"Тип результата: {type(all_phones)}")
print(f"Количество найденных телефонов: {len(all_phones)}")
````

### `re.split()`: разбиение по шаблону

Стандартный метод строки `.split()` отлично разбивает строку по фиксированному разделителю (или по пробелам). Но что если разделители могут быть разными или представлять собой более сложный шаблон? Здесь на помощь приходит `re.split()`.

*   `re.split(pattern, string, maxsplit=0)`: Разбивает строку `string` по **совпадениям** с шаблоном `pattern`.
*   Возвращает **список строк**, полученных после разбиения. Сами разделители (совпадения с `pattern`) в результат не попадают.
*   `maxsplit` (необязательный): Максимальное количество разбиений.

**Отличие от `string.split()`:** `re.split()` использует **регулярное выражение** в качестве разделителя.

````python
data_string = "яблоки, бананы; груши   апельсины:манго"

# Хотим разбить по ЛЮБОМУ из разделителей: ',', ';', ':', ИЛИ по одному или нескольким пробелам '\s+'
# Используем символьный класс [] и метасимвол \s+
split_pattern = r'[;,:\s]+' # Один или более символов из набора: ;, :, или пробельный символ

items = re.split(split_pattern, data_string)
print(f"Строка: '{data_string}'")
print(f"Результат re.split(): {items}")

# Сравните с обычным split, который так не сможет:
print(f"Результат data_string.split(','): {data_string.split(', ')}")
````

`re.split()` незаменим, когда нужно разобрать строку со сложными или множественными разделителями.




### Группы захвата `()`: извлекаем части совпадения

Часто нам нужно не просто найти совпадение с шаблоном, но и **извлечь из этого совпадения отдельные фрагменты**. Например, из строки с датой "2023-10-27" извлечь отдельно год, месяц и день.

Для этого используются **круглые скобки `()`** в шаблоне. Часть шаблона, заключенная в скобки, называется **группой захвата (capturing group)**.

**Как это работает:**

*   Когда шаблон с группами находит совпадение, движок regex не только запоминает всё совпадение целиком (доступное через `.group()` или `.group(0)`), но и **запоминает текст, совпавший с каждой группой** по отдельности.
*   Доступ к захваченным группам можно получить по их **порядковому номеру** (начиная с 1) с помощью метода `.group(номер_группы)`.

**Пример: Разбор даты**

````python
date_string = "Сегодняшняя дата: 2025-10-27"
# Шаблон с группами для года, месяца и дня
# Группа 1: (\d{4}) - 4 цифры (год)
# Группа 2: (\d{2}) - 2 цифры (месяц)
# Группа 3: (\d{2}) - 2 цифры (день)
date_pattern_groups = r'(\d{4})-(\d{2})-(\d{2})'

match_ = re.search(date_pattern_groups, date_string)

if match_:
    print(match_)
    # Полное совпадение (группа 0 - это всегда всё совпадение)
    full_match = match_.group() # или просто .group()
    print(f"Полное совпадение (group 0): '{full_match}'")

    # Извлекаем отдельные группы по номерам
    year = match_.group(1)
    month = match_.group(2)
    day = match_.group(3)
    print(f"Год (group 1): {year}")
    print(f"Месяц (group 2): {month}")
    print(f"День (group 3): {day}")

    # Метод .groups() возвращает кортеж со всеми захваченными группами (начиная с 1)
    all_groups_tuple = match_.groups()
    print(f"Все группы кортежем (.groups()): {all_groups_tuple}")
else:
    print(f"Дата в формате YYYY-MM-DD не найдена в строке.")
````

**Важно:**
*   Нумерация групп идет слева направо по **открывающим** скобкам.
*   Группа 0 – это всегда всё совпадение.
*   Если группа не нашла совпадения (например, из-за квантификатора `?` или `*`), `.group(номер)` вернет `None`.



**Пример: Разбор Email**
Извлечем имя пользователя и домен, дав им осмысленные имена.

````python
email = "support.team@my-company.com"
# Шаблон с именованными группами 'user' и 'domain'
# [\w.-]+ - имя пользователя (буквы, цифры, _, ., -)
# [\w.-]+ - домен (такой же набор символов, для упрощения)
email_pattern_named = r'([\w.-]+)@([\w.-]+)'

match_ = re.search(email_pattern_named, email)

if match_:
    print(f"Полное совпадение: '{match_.group()}'")
    # Доступ по именам
    username = match_.group(1)
    domain_name = match_.group(2)
    print(f"Имя пользователя ('user'): {username}")
    print(f"Домен ('domain'): {domain_name}")
else:
    print("Email не найден.")
````

**Группы и `re.findall()`:**

Если шаблон в `findall` содержит группы, результат меняется:
*   Если **одна** группа: `findall` вернет список строк, захваченных этой группой.
*   Если **несколько** групп: `findall` вернет список **кортежей**, где каждый кортеж содержит строки, захваченные всеми группами для одного совпадения.

````python
text_with_pairs = "Координаты: (10, 20), (30, -5), (0, 0)."
# Шаблон для поиска пар чисел в скобках, с 2 группами для каждого числа
pair_pattern = r'\((-?\d+),\s*(-?\d+)\)' # Группа 1: первое число, Группа 2: второе число

# findall вернет список кортежей ('число1', 'число2')
list_of_pairs = re.findall(pair_pattern, text_with_pairs)

if list_of_pairs:
    print(list_of_pairs)
    print("Распакованные пары:")
    for x, y in list_of_pairs: # Распаковываем кортеж
        print(f"  x={x}, y={y}")

````

### Использование групп в `re.sub()`

Функция `re.sub(pattern, replacement, string)` заменяет все вхождения `pattern` на `replacement`. Но самое интересное начинается, когда `pattern` содержит группы захвата, а строка `replacement` на них ссылается!



**Пример 1: Преобразование формата даты (Нумерованные группы)**
Задача: Преобразовать даты YYYY-MM-DD в формат DD.MM.YYYY.

````python
date_text_iso = "Дата отчета: 2025-10-27, крайний срок: 2026-01-15."

# Шаблон с 3 нумерованными группами: (год)-(месяц)-(день)
iso_date_pattern = r'(\d{4})-(\d{2})-(\d{2})'
#                   | Гр.1 | | Гр.2| | Гр.3|
# Строка замены: используем \3 (день), \2 (месяц), \1 (год)
replacement_format_dmY = r'\3.\2.\1' # День.Месяц.Год

# Выполняем замену
date_text_dmY = re.sub(iso_date_pattern, replacement_format_dmY, date_text_iso)

print(f"Исходный текст: {date_text_iso}")
print(f"Текст с новым форматом дат: {date_text_dmY}")
````

**Пример 2: Перестановка имени и фамилии**  
Задача: В тексте вида "Иванов Петр" поменять местами фамилию и имя, добавив запятую: "Петр Иванов".

````python
name_list_text = "Контакты: Иванов Петр, Сидорова Анна, Кузнецов Олег."

# Используем \w+ - одна или более букв/цифр/_
name_pattern_named = r'([А-Яа-яЁё]+)\s+([А-Яа-яЁё]+)'

replacement_format_name = r'\2 \1'

# Выполняем замену
name_list_reordered = re.sub(name_pattern_named, replacement_format_name, name_list_text)

print(f"Исходный текст: {name_list_text}")
print(f"Текст после перестановки: {name_list_reordered}")
````
